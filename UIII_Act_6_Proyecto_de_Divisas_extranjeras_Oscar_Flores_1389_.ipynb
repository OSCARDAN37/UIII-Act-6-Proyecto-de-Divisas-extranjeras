```python

print("==========================================")
print("🤖 ASISTENTE DE VIAJES INTELIGENTE 🤖")
print("==========================================")

continuar = "si"

while continuar == "si":
    print("\n--- MENÚ DE DESTINOS ---")
    print("1. Viajar a Europa (Euros - EUR)")
    print("2. Viajar a México (Pesos - MXN)")
    print("3. Salir del asistente")

    opcion = input("Selecciona tu destino deseado (1, 2 o 3): ")

    if opcion == "1":
        dolares = float(input("¿Cuántos dólares (USD) tienes para tu viaje?: "))
        euros = dolares * 0.92
        print("Tu dinero equivale a:", euros, "Euros.")

        if euros < 800:
            print("🤖 IA Dice: Presupuesto bajo para Europa. Te sugiero ahorrar un poco más.")
        else:
            print("🤖 IA Dice: ¡Presupuesto excelente! Tienes suficiente para disfrutar Europa.")

    elif opcion == "2":
        dolares = float(input("¿Cuántos dólares (USD) tienes para tu viaje?: "))
        pesos = dolares * 16.80
        print("Tu dinero equivale a:", pesos, "Pesos Mexicanos.")

        if pesos < 10000:
            print("🤖 IA Dice: Presupuesto ajustado. Deberás buscar hoteles económicos.")
        else:
            print("🤖 IA Dice: ¡Presupuesto genial! Podrás pasar unas vacaciones muy cómodas.")

    elif opcion == "3":
        print("🤖 Asistente apagado. ¡Buen viaje!")
        continuar = "no"

    else:
        print("❌ Opción incorrecta. Elige 1, 2 o 3.")

    if opcion != "3":
        continuar = input("\n¿Quieres consultar otro destino? (si/no): ")

print("==========================================")
print("Programa realizado por Oscar Flores NL = 22")
```
